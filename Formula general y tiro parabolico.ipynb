```python
%matplotlib notebook 
#Tenia la falla "<Figure size 640x480 with 1 Axes>", con este codigo se arreglo el problema
import matplotlib.pyplot as plt
import numpy as np
print("Este programa con los coeficientes de La formula general, determina si la ecuación cuadratica tiene dos raices reales y diferentes, sus dos raices son iguales o si son imaginarias e imprime su grafica")
print("La formula general es: x=(-b±√(b²-4ac))/(2a) , y la ecuacion cuadratica es: ax²+bx+c")
a=float(input("Ingrese (a): "))
while a == 0:
    a=input("El termino cuadratico no puede ser 0, Ingrese (a): ")
b=float(input("Ingrese (b): "))
c=float(input("Ingrese (c): "))
print ("a continuacion, ingresara desde que valor en x y hasta que valor desea imprimir la grafica")
xi=float(input("Ingrese desde donde empeiza la grafica: "))
xf=float(input("Ingrese hasta donde termina la grafica: "))
while xf<=xi:
        if xf < xi:
            #xf es donde termina de graficar y xi donde empieza, no pueden ser iguales y xf no puede ser menor que el inicial
            print ("Donde finaliza no puede ser menor a donde termina") #Para darle mas presentacion, se imprimira un mensaje personalizado donde se vera si es igual o menor xf
        else:
             print ("Donde empieza no puede ser igual a donde termina")
        print ("Ingrese hasta donde termina la grafica: ")
        xf=float(input())
xs=(xf-xi)/50 #xs, seran los saltos para la grafica, si los saltos son muy grandes, de 1 o mas, se le asignara un 0.1 en los saltos, si el usuario da un rango muy pequeño para la grafica, los saltos de 0.1, seran muy grandes, por eso se divide en 100
if xs>=1:
         xs=0.1
d=float(b)**2-4*a*c #con la discriminante podemos encontrar la naturaleza de sus raices
#primero, se imprimira la naturaleza de las raices
if d == 0: #si la discriminante es 0 nos daran raices iguales, por lo tanto solo ocupamos imprimir una raiz
    print("Las dos raices son reales e iguales")
    x1=-b/(2*a)
    print("r="+str(x1))
    f="s"
elif d < 0: #si la discriminante es menor a 0,las raices seran complejas
        print("Las dos raices son complejas")
        x1=-b/(2*a)
        x2=((-d)**0.5)/2*a #al ser la discriminante negativa, la multiplicamos por -1 para poder obtener la raiz
        print("r1= "+str(x1)+"+"+str(x2)+"i")
        print("r2= "+str(x1)+"-"+str(x2)+"i")
        f="n"
elif d > 0: #la raiz es positiva, por lo tanto la raiz se podra efectuar sin problemas y nos dara un resultado de raiz real y diferente
        print("Las dos raices son reales y diferentes")
        x1=(-b+d**0.5)/(2*a)
        x2=(-b-d**0.5)/(2*a)
        print("r1= "+str(x1))
        print("r2= "+str(x2))
        f="m"
#seguido, se imprimira la grafica
x=np.arange(xi,xf,xs)
y=a*x**2+b*x+c #Es la forma ax²±bx+c
plt.plot(x,y)
if f=="s": #Si solo tiene una raiz, solo imprimira una marca
    y1=a*x1**2+b*x1+c
    plt.plot(x1,y1, 'm*')
elif f=="m": #Si tiene 2 raices, imprimira 2 marcas
    y1=a*x1**2+b*x1+c
    y2=a*x2**2+b*x2+c
    plt.plot(x1,y1, 'm*')
    plt.plot(x2,y2, 'r*')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Ecuación cuadratica")

plt.show()
```
